```python
import mediapipe as mp
import cv2
import os
from PIL import Image
from tqdm import tqdm

base_dir = "/content/drive/MyDrive/photo-restoration-lora/data"

color_dir = f"{base_dir}/raw/color"
clean_output_dir = f"{base_dir}/clean"
os.makedirs(clean_output_dir, exist_ok=True)
```

```python
mp_face = mp.solutions.face_detection
detector = mp_face.FaceDetection(model_selection=1, min_detection_confidence=0.5)

print("Processing ONLY color images...")

for fname in tqdm(os.listdir(color_dir)):
    if not fname.lower().endswith((".jpg", ".jpeg", ".png")):
        continue

    img_path = os.path.join(color_dir, fname)
    img = cv2.imread(img_path)

    if img is None:
        print("Could not load:", img_path)
        continue

    h, w, _ = img.shape

    # Convert to RGB for Mediapipe
    rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    results = detector.process(rgb)

    if not results.detections:
        print("Face not detected — skipping:", img_path)
        continue

    # Take the first detected face
    detection = results.detections[0]
    bbox = detection.location_data.relative_bounding_box

    x1 = int(bbox.xmin * w)
    y1 = int(bbox.ymin * h)
    x2 = int
```
